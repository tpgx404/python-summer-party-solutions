## Question 1

What is the average click-through rate (CTR) for sponsored product ads for each product category that contains the substring 'Electronics' in its name during October 2024? This analysis will help determine which electronics-related categories are performing optimally.

```python
# Note: pandas and numpy are already imported as pd and np
# The following tables are loaded as pandas DataFrames with the same names: fct_ad_performance, dim_product
# Please print your final result or dataframe
# 1. Convert recorded_date to datetime for filtering
fct_ad_performance['recorded_date'] = pd.to_datetime(fct_ad_performance['recorded_date'])

# 2. Filter for October 2024
october_ads = fct_ad_performance[
    (fct_ad_performance['recorded_date'].dt.year == 2024) &
    (fct_ad_performance['recorded_date'].dt.month == 10)
]

# 3. Merge with dim_product to get category info
ads_with_category = october_ads.merge(dim_product, on='product_id', how='left')

# 4. Filter categories containing 'Electronics'
electronics_ads = ads_with_category[ads_with_category['product_category'].str.contains('Electronics', case=False)]

# 5. Calculate CTR per row
electronics_ads['CTR'] = (electronics_ads['clicks'] / electronics_ads['impressions']) * 100

# 6. Group by category and calculate average CTR
avg_ctr_by_category = electronics_ads.groupby('product_category')['CTR'].mean().reset_index()

print(avg_ctr_by_category)
```

## Question 2

Which product categories have a CTR greater than the aggregated overall average CTR for sponsored product ads during October 2024? This analysis will identify high-performing categories for further optimization. For this question, we want to calculate CTR for each ad, then get the average across ads by product category & overall.

```python
# 1. Convert date for filtering
fct_ad_performance['recorded_date'] = pd.to_datetime(fct_ad_performance['recorded_date'])

# 2. Filter October 2024
october_ads = fct_ad_performance[
    (fct_ad_performance['recorded_date'].dt.year == 2024) &
    (fct_ad_performance['recorded_date'].dt.month == 10)
]

# 3. Merge with product info
ads_with_category = october_ads.merge(dim_product, on='product_id', how='left')

# 4. Calculate CTR for each ad
ads_with_category['CTR'] = (ads_with_category['clicks'] / ads_with_category['impressions']) * 100

# 5. Average CTR per product category
category_ctr = ads_with_category.groupby('product_category')['CTR'].mean().reset_index()

# 6. Overall average CTR
overall_avg_ctr = ads_with_category['CTR'].mean()

# 7. Filter categories with CTR > overall average
high_performing_categories = category_ctr[category_ctr['CTR'] > overall_avg_ctr]

print(high_performing_categories)
```

## Question 3

For the product categories identified in the previous question, what is the percentage difference between their CTR and the overall average CTR for October 2024? This analysis will quantify the performance gap to recommend specific categories for targeted advertising optimization.

```python
# 1. Convert dates for filtering
fct_ad_performance['recorded_date'] = pd.to_datetime(fct_ad_performance['recorded_date'])

# 2. Filter for October 2024
october_ads = fct_ad_performance[
    (fct_ad_performance['recorded_date'].dt.year == 2024) &
    (fct_ad_performance['recorded_date'].dt.month == 10)
]

# 3. Merge with product info
ads_with_category = october_ads.merge(dim_product, on='product_id', how='left')

# 4. Calculate CTR for each ad
ads_with_category['CTR'] = (ads_with_category['clicks'] / ads_with_category['impressions']) * 100

# 5. Compute average CTR per product category
category_ctr = ads_with_category.groupby('product_category')['CTR'].mean().reset_index()

# 6. Compute overall average CTR
overall_avg_ctr = ads_with_category['CTR'].mean()

# 7. Filter high-performing categories (CTR > overall avg)
high_performing_categories = category_ctr[category_ctr['CTR'] > overall_avg_ctr].copy()

# 8. Compute percentage difference for these categories
high_performing_categories['percentage_diff'] = (
    (high_performing_categories['CTR'] - overall_avg_ctr) / overall_avg_ctr
) * 100

# 9. Optional: sort by performance gap
high_performing_categories = high_performing_categories.sort_values('percentage_diff', ascending=False)

print(high_performing_categories)
```
