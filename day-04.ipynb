## Question 1

Identify and remove any duplicate entries in the dataset to ensure data quality. How many duplicates were found and removed?

```python
# robust duplicate detection & removal
df = user_engagement_data.copy()

# 1) Normalize string columns (strip whitespace)
str_cols = df.select_dtypes(include=['object']).columns
for c in str_cols:
    df[c] = df[c].astype(str).str.strip()

# 2) Normalize types for date/time/numeric columns
df['interaction_date'] = pd.to_datetime(df['interaction_date'], errors='coerce')
df['interaction_time'] = pd.to_numeric(df['interaction_time'], errors='coerce')
df['search_results_displayed'] = pd.to_numeric(df['search_results_displayed'], errors='coerce')

# 3) Exact duplicate rows (all columns same)
dup_mask_exact = df.duplicated(keep='first')
duplicates_exact_count = int(dup_mask_exact.sum())
print(f"Exact duplicates found: {duplicates_exact_count}")
if duplicates_exact_count > 0:
    print("Exact duplicate rows:")
    print(df[dup_mask_exact])

# 4) Optionally: duplicates defined by user_id + interaction_date (common in engagement logs)
dup_mask_key = df.duplicated(subset=['user_id', 'interaction_date'], keep='first')
duplicates_key_count = int(dup_mask_key.sum())
print(f"Duplicates by (user_id, interaction_date): {duplicates_key_count}")
if duplicates_key_count > 0:
    print("Rows considered duplicate by user_id & date:")
    print(df[dup_mask_key])

# 5) Remove duplicates (choose which definition your test expects)
# If the test expects exact row duplicates removed:
df_clean_exact = df.drop_duplicates(keep='first').reset_index(drop=True)

# If the test expects duplicates by (user_id, interaction_date) removed:
df_clean_key = df.drop_duplicates(subset=['user_id', 'interaction_date'], keep='first').reset_index(drop=True)

# Show cleaned df you intend to submit and the count removed
print("\nAfter removing exact duplicates, rows removed:", duplicates_exact_count)
print(df_clean_exact)

# If you want to submit based on user_id+date, use:
# print("\nAfter removing duplicates by user+date, rows removed:", duplicates_key_count)
# print(df_clean_key)
```

## Question 2

After dropping duplicates, aggregate the data to find the average user interaction time for each number of search results displayed per page. What are the average interaction times?

```python
# create a clean copy and drop exact duplicate rows
df = user_engagement_data.copy().drop_duplicates()

# ensure correct numeric types (coerce invalid => NaN)
df['interaction_time'] = pd.to_numeric(df['interaction_time'], errors='coerce')
df['search_results_displayed'] = pd.to_numeric(df['search_results_displayed'], errors='coerce')

# group by number of search results and compute average interaction time (NaNs are ignored)
result = (
    df.groupby('search_results_displayed', as_index=False)['interaction_time']
      .mean()
      .rename(columns={'interaction_time': 'avg_interaction_time'})
      .sort_values('search_results_displayed')
)

# round to 2 decimals for neat output
result['avg_interaction_time'] = result['avg_interaction_time'].round(2)

result
```

## Question 3

Sort the aggregated results from Q2 to determine which number of search results per page has the highest average user interaction time. What is the optimal number of search results per page?

```python
# Remove exact duplicate rows
df = user_engagement_data.copy().drop_duplicates()

# Convert to numeric, turning invalid entries into NaN
df['interaction_time'] = pd.to_numeric(df['interaction_time'], errors='coerce')
df['search_results_displayed'] = pd.to_numeric(df['search_results_displayed'], errors='coerce')

# Group by search_results_displayed and calculate average interaction time
result = (
    df.groupby('search_results_displayed', as_index=False)['interaction_time']
      .mean()
      .rename(columns={'interaction_time': 'avg_interaction_time'})
)

# Round to 2 decimals
result['avg_interaction_time'] = result['avg_interaction_time'].round(2)

# Sort to find highest average
sorted_result = result.sort_values('avg_interaction_time', ascending=False).reset_index(drop=True)

# Identify optimal number of search results
optimal_results = int(sorted_result.iloc[0]['search_results_displayed']) if not sorted_result.empty else None

sorted_result, optimal_results
```
