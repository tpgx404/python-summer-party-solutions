## Question 2

Using the cleaned data, calculate the total pre-sale orders per month for each region and demographic group.

```python
import pandas as pd

# Use cleaned data if available; otherwise create it by dropping rows with missing values
if 'pre_sale_data_cleaned' in globals():
    df = pre_sale_data_cleaned.copy()
else:
    df = pre_sale_data.copy().dropna()

# Ensure the pre_order_date column is datetime
df['pre_order_date'] = pd.to_datetime(df['pre_order_date'], errors='coerce')

# Extract month in YYYY-MM format for grouping
df['month'] = df['pre_order_date'].dt.strftime('%Y-%m')

# Group by month, region, and demographic_group and sum pre_order_quantity
monthly_orders = (
    df.groupby(['month', 'region', 'demographic_group'], as_index=False)['pre_order_quantity']
      .sum()
)

# Sort for readability (by month, then by descending orders)
monthly_orders = monthly_orders.sort_values(['month', 'pre_order_quantity'], ascending=[True, False]).reset_index(drop=True)

monthly_orders
```

## Question 3

Predict the total pre-sales quantity for each region for September 2024. Assume that growth rate from August to September, is the same as the growth rate from July to August in each region.

```python
import pandas as pd

# Use cleaned data if present; otherwise fall back to dropping rows with missing values
if 'pre_sale_data_cleaned' in globals():
    df = pre_sale_data_cleaned.copy()
else:
    df = pre_sale_data.copy().dropna()

# Ensure date column is datetime and create YYYY-MM month key
df['pre_order_date'] = pd.to_datetime(df['pre_order_date'], errors='coerce')
df['month'] = df['pre_order_date'].dt.strftime('%Y-%m')

# Aggregate total pre-orders per region and month
monthly_region_totals = df.groupby(['month', 'region'], as_index=False)['pre_order_quantity'].sum()

# Pivot so regions are rows and months are columns, fill missing with 0
pivot = monthly_region_totals.pivot(index='region', columns='month', values='pre_order_quantity').fillna(0)

# Ensure July and August columns exist
for m in ['2024-07', '2024-08']:
    if m not in pivot.columns:
        pivot[m] = 0.0

# Calculate growth rate safely: (Aug - Jul) / Jul for Jul > 0, else leave growth_rate = 0
pivot['growth_rate'] = 0.0
mask = pivot['2024-07'] > 0
pivot.loc[mask, 'growth_rate'] = (pivot.loc[mask, '2024-08'] - pivot.loc[mask, '2024-07']) / pivot.loc[mask, '2024-07']

# Predict September = August * (1 + growth_rate)
pivot['predicted_2024_09'] = pivot['2024-08'] * (1 + pivot['growth_rate'])

# Prepare final results, rounding to nearest integer
predictions = pivot[['predicted_2024_09']].reset_index().rename(columns={'predicted_2024_09': 'predicted_pre_order_quantity'})
predictions['predicted_pre_order_quantity'] = predictions['predicted_pre_order_quantity'].round().astype(int)

predictions
```
