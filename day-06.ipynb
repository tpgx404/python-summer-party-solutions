## Question 2

Create a pivot table to summarize the total sales volume of ice cream products by month and temperature range.
Use the following temperature bins where each bin excludes the upper bound but includes the lower bound:
- Less than 60 degrees
- 60 to less than 70 degrees
- 70 to less than 80 degrees
- 80 to less than 90 degrees
- 90 to less than 100 degrees
- 100 degrees or more

```python
import pandas as pd

# Sample data (replace with actual loading code)
data = {
    "sale_date": [
        "2024-07-05","2024-08-15","2024-09-25","2024-10-05","2024-11-15",
        "2024-12-25","2025-01-05","2025-02-15","2025-03-25","2025-04-05","2025-05-15"
    ],
    "temperature": [62,64,66,68,70,72,74,76,78,80,82],
    "product_name": [
        "Cherry Garcia","Chunky Monkey","Phish Food","Americone Dream",
        "Chocolate Fudge Brownie","Half Baked","New York Super Fudge Chunk",
        "Cherry Garcia","Chunky Monkey","Phish Food","Americone Dream"
    ],
    "sales_volume": [23,26,29,32,35,38,41,44,47,50,53],
    "transaction_id": [
        "TX0001","TX0002","TX0003","TX0004","TX0005","TX0006",
        "TX0007","TX0008","TX0009","TX0010","TX0011"
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# Ensure sale_date is datetime
df['sale_date'] = pd.to_datetime(df['sale_date'])

# Extract month name
df['month'] = df['sale_date'].dt.strftime('%B')

# Define temperature bins and labels
bins = [-float('inf'), 60, 70, 80, 90, 100, float('inf')]
labels = [
    "Less than 60 degrees",
    "60 to less than 70 degrees",
    "70 to less than 80 degrees",
    "80 to less than 90 degrees",
    "90 to less than 100 degrees",
    "100 degrees or more"
]

# Create temperature range category
df['temp_range'] = pd.cut(df['temperature'], bins=bins, labels=labels, right=False)

# Create pivot table
pivot_table = pd.pivot_table(
    df,
    values='sales_volume',
    index='month',
    columns='temp_range',
    aggfunc='sum',
    fill_value=0
)

print(pivot_table)
```

## Question 3

Can you detect any outliers in the monthly sales volume using the Inter Quartile Range (IQR) method? A month is considered an outlier if falls below Q1 minus 1.5 times the IQR or above Q3 plus 1.5 times the IQR.

```python
import pandas as pd

# Sample data
data = {
    "sale_date": [
        "2024-07-05","2024-08-15","2024-09-25","2024-10-05","2024-11-15",
        "2024-12-25","2025-01-05","2025-02-15","2025-03-25","2025-04-05","2025-05-15"
    ],
    "temperature": [62,64,66,68,70,72,74,76,78,80,82],
    "product_name": [
        "Cherry Garcia","Chunky Monkey","Phish Food","Americone Dream",
        "Chocolate Fudge Brownie","Half Baked","New York Super Fudge Chunk",
        "Cherry Garcia","Chunky Monkey","Phish Food","Americone Dream"
    ],
    "sales_volume": [23,26,29,32,35,38,41,44,47,50,53],
    "transaction_id": [
        "TX0001","TX0002","TX0003","TX0004","TX0005","TX0006",
        "TX0007","TX0008","TX0009","TX0010","TX0011"
    ]
}

# Create DataFrame
df = pd.DataFrame(data)

# Ensure sale_date is datetime
df['sale_date'] = pd.to_datetime(df['sale_date'])

# Extract month name
df['month'] = df['sale_date'].dt.strftime('%B')

# Calculate total monthly sales
monthly_sales = df.groupby('month')['sales_volume'].sum()

# Calculate Q1, Q3, and IQR
Q1 = monthly_sales.quantile(0.25)
Q3 = monthly_sales.quantile(0.75)
IQR = Q3 - Q1

# Determine outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Find outlier months
outliers = monthly_sales[(monthly_sales < lower_bound) | (monthly_sales > upper_bound)]

print("Monthly Sales:")
print(monthly_sales)
print("\nQ1:", Q1, " Q3:", Q3, " IQR:", IQR)
print("Lower Bound:", lower_bound, " Upper Bound:", upper_bound)
print("\nOutlier Months:")
print(outliers)
```
