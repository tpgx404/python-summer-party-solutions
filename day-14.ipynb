## Question 2

What is the average transaction value for loyalty program members and non-members during July 2024? Use this to identify which group has a higher average transaction value.

```python
import pandas as pd

# Assuming your tables are loaded as DataFrames
# fct_transactions: transaction_id, customer_id, transaction_date, transaction_value
# dim_customers: customer_id, is_loyalty_member

# 1. Convert transaction_date to datetime
fct_transactions['transaction_date'] = pd.to_datetime(fct_transactions['transaction_date'])

# 2. Filter for July 2024 transactions
july_txns = fct_transactions[
    (fct_transactions['transaction_date'].dt.year == 2024) &
    (fct_transactions['transaction_date'].dt.month == 7)
]

# 3. Join with customers table to get loyalty info
merged = july_txns.merge(dim_customers, on="customer_id", how="left")

# 4. Group by loyalty membership and calculate average transaction value
avg_values = merged.groupby("is_loyalty_member")['transaction_value'].mean().reset_index()

# 5. Rename columns for clarity
avg_values.columns = ['is_loyalty_member', 'avg_transaction_value']

print(avg_values)
```

## Question 3

Determine the percentage difference in average transaction value between loyalty program members and non-members for July 2024.

```python
import pandas as pd

# 1. Ensure datetime
fct_transactions['transaction_date'] = pd.to_datetime(fct_transactions['transaction_date'])

# 2. Filter July 2024 transactions
july_txns = fct_transactions[
    (fct_transactions['transaction_date'].dt.year == 2024) &
    (fct_transactions['transaction_date'].dt.month == 7)
]

# 3. Merge with customer info
merged = july_txns.merge(dim_customers, on="customer_id", how="left")

# 4. Standardize loyalty column (convert to Yes/No)
merged['is_loyalty_member'] = merged['is_loyalty_member'].astype(str).str.lower().map({
    '1': 'Yes', 'true': 'Yes', 'yes': 'Yes', 
    '0': 'No', 'false': 'No', 'no': 'No'
})

# 5. Calculate average transaction value by loyalty status
avg_values = merged.groupby("is_loyalty_member")['transaction_value'].mean()

# 6. Extract values safely
member_avg = avg_values.get('Yes', 0)
nonmember_avg = avg_values.get('No', 0)

# 7. Percentage difference (relative to non-members)
if nonmember_avg != 0:
    pct_diff = ((member_avg - nonmember_avg) / nonmember_avg) * 100
else:
    pct_diff = None

print("Loyalty Members Avg:", round(member_avg, 2))
print("Non-Members Avg:", round(nonmember_avg, 2))
print("Percentage Difference:", round(pct_diff, 2), "%")
```
